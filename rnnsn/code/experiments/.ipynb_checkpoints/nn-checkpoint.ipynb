```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
seed = 42
np.random.seed(seed)
```

```python
X, y = getAggrSet()
X_train, X_test, y_train, y_test = splitAndNormaliseAggr(X,y)
y_train = y_train.values
y_test = y_test.values
```

```python
X.shape[1]
```

```python
def baseline_model():
    model = Sequential()
    model.add(Dense(X.shape[1], input_dim=X.shape[1], init='normal', activation='relu'))
    model.add(Dense(1, init='normal'))
    model.compile(loss='mean_squared_error', optimizer='adam')
    return model
```

```python
estimator = KerasRegressor(build_fn=baseline_model, nb_epoch=10, batch_size=5, verbose=0)
```

```python
kfold = KFold(n_splits=10, random_state=seed)
results = cross_val_score(estimator, X_train, y_train, cv=kfold)
```

```python
print("Results: %.2f (%.2f) MSE" % (results.mean(), results.std()))
```
