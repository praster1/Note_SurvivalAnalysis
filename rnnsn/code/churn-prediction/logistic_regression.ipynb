```python
from churn_data import ChurnData
```

```python
#matplotlib inline
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectFromModel
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = ChurnData()
```

# Train model

```python
model = LogisticRegression(penalty='l1')
model = model.fit(**data.train)
```

```python
data.getScore(model)
```

```python
featureSelection = SelectFromModel(model, prefit=True)
```

```python
pd.DataFrame(list(zip(data.features, model.coef_.reshape(-1), np.abs(model.coef_.reshape(-1)), featureSelection.get_support()))).sort_values(by=2, ascending=False)
```
